```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

### extrac batting performance

```python
#The first link all about most runs getter in the series 
#I used espn cricinfo website link

req=requests.get("https://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?id=15029;type=tournament")
#with the help of request i get html  data 
soup=BeautifulSoup(req.content,"html.parser")
#with the help of bs4 i parser the html file
table=soup.table # table is used here for store the table content
#print(table) if you to see that please print that
thead=table.find("thead").find_all('th') #using find function for find table head 
print(thead)

```

```python
table_head= [th.text for th in thead]     #using for loop to extrac the table head put that into list
print(table_head)
```

```python
table_body = table.find("tbody").find_all("tr")    #here i am using find for find table body
print(table_body)
```

```python
for tr in table_body:                         #i used for loop to extrac content rowise and put that in list
    table_data=[  td.text.strip() for td in tr.find_all("td")]  
    print(table_data)
```

```python
#here i am using csv to creat a report and store the table head and table body what i extra from website


with open("player_most_run_report(1).csv","w",newline="") as csv_file:
    csv_write=csv.writer(csv_file)
    csv_write.writerow(table_head)
    
    for tr in table_body:
        table_data=[  td.text.strip() for td in tr.find_all("td")]  
        csv_write.writerow(table_data)
```
